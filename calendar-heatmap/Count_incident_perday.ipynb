```python
import pandas as pd
import numpy as np
```

```python
data_path = 'subset_data.csv'
data = pd.read_csv(data_path)
data.head()
```

```python
main_df = data.drop(['year', 'time', 'category', 'meta_cat', 'location', 'zipcode', 
                        'lat', 'lng', 'resolution', 'neighborhood'], axis=1)
data_bymonth = main_df.groupby(['day_m', 'month'])
month_day_df = data_bymonth["id"].size()
month_day_df = month_day_df.unstack()
month_day_df
```

```python
month_day_df.to_csv('incident_perday.csv', index=True, header=True)
```
